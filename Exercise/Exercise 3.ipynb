### 5. Is there a size difference between the initial JPG and the bitmap file? Why?
Yes, there is a size difference between a JPG file and a bitmap (BMP) file. The BMP format is an uncompressed image format which means it stores color data for each pixel in the image without any compression. As a result, the file size of a BMP image can be larger than that of a JPG image.

On the other hand, JPG is a lossy compression format. It reduces file size by selectively discarding certain information that is less noticeable to human eyes. This is why a JPG image is usually smaller in size compared to a BMP image of the same picture.

The bmp file is 2367KB, on the other hand jpg file is 853KB

### 6. Which one is better in terms of image quality?
In terms of image quality, the BMP format is technically better because it is an uncompressed format. This means it stores all the original color data for each pixel in the image without any loss of information. As a result, the image quality of a BMP file is typically higher than that of a JPG file.

## Exercise 3.1.3 Image structure

### 1. How is the image from the previous exercise stored/represented in Python? What’s its data type?

> When we load an image using skimage.data.coins(), the image is loaded into a 2D NumPy array when it’s a grayscale image. Each element in the array represents a pixel in the image. The value of each pixel can range from 0 (black) to 255 (white) for an 8-bit grayscale image.

### 2. What are the dimensions of the image?

```python
print('The dimension of the image is :', img1.shape)
```

### 3. What are the possible values of each entry?

Each entry in the image array represents a pixel intensity and for an 8-bit grayscale image, each entry is an integer that ranges from 0 to 255. Here's what these values represent:

- **0** represents black.
- **255** represents white.
- Values in between represent various shades of gray, where lower numbers are darker and higher numbers are lighter.
